```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:/Courses/Machine Learning\Market.csv')
data
```

```python
df=pd.DataFrame(data)
df.dtypes
```

```python
df1=df.loc[(df['Index']=='NYA')]
df1
```

```python

df2=pd.to_datetime(df1['Date'])
df2.columns = ['index','Date']
print(df2)
```

```python
df.dtypes
```

```python
plt.plot(df2,df1['Adj Close'])
plt.ylabel('Adj Close')
plt.xlabel('Date')

plt.savefig('Date vs adj.jpg')
plt.show()
```
